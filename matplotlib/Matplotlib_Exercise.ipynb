  (7) 임의의 함수를 정하고 이를 미분한 함수를 같이 만드세요. 이 둘을 한개의 그래프에 같이 표시하고 미분값이 의미하는 바를 적어보세요.

```python
"""
이 셀 안에 코드를 작성하시오.
"""
x = np.linspace(-4,4,50)
y1 = x**3 - 9*x
y2 = 3*(x**2)-18

plt.plot(x,y1,linestyle = '-',color = 'red',label = "$y = f(x)$")
plt.plot(x,y2,linestyle = '--',color = 'blue',label = "y = f'(x)")
plt.legend(shadow = True, borderpad = 1)
plt.show()
```
